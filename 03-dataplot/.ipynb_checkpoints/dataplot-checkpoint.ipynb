```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('winemag-data-130k-v2.csv', index_col='Unnamed: 0')
```

```python
df = df.loc[df['country']=='US']
df
```

```python
df = df[pd.notnull(df['price'])]
df = df[pd.notnull(df['points'])]
```

```python
def floattoint(i):
    return int(i)
df["price"] = df.apply(lambda row: floattoint(row["price"]), axis=1)
df
```

```python
x = list(df['price'])
y = list(df['points'])
```

```python
plt.scatter(x, y)
plt.xlabel('Price')
plt.ylabel('Points')
plt.title('Price v Points for US Wines')
plt.show()
```

```python
df = df.loc[df['price']<2000]
x = list(df['price'])
y = list(df['points'])
plt.scatter(x, y)
plt.xlabel('Price')
plt.ylabel('Points')
plt.title('Price v Points for US Wines')
plt.show()
```
